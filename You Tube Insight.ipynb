## 15) Which Country has highest unemployemment rate ?

# 17) How do various metrics or activity correlate, providing insights into their interplay and potential impact on YouTube channel performance ?

```python
# Update relevant variables list to exclude non-numeric columns
relevant_variables = ['video views', 'subscribers', 'Population', 'Unemployment rate', 'Urban_population',
                      'subscribers_for_last_30_days']

# Subset the DataFrame to include only the relevant variables
relevant_df = df[relevant_variables]

# Calculate the correlation matrix
correlation_matrix = relevant_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Relevant Variables')
plt.show()
```

### There exists a strong positive correlation between subscriber count and video views, as evidenced by the histogram plot, suggesting that channels with higher subscriber counts tend to garner more views.

# 21) Which country experienced the highest growth in highest monthly earnings over the observed years

```python
# Aggregate the data by country and created year, taking the sum of highest yearly earnings
agg_df = df.groupby(['Country', 'created_year'])['highest_yearly_earnings'].sum().reset_index()

# Plot the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=agg_df, x='created_year', y='highest_yearly_earnings', hue='Country', marker='o')
plt.title('Highest Yearly Earning by Created Year and Country')
plt.xlabel('Created Year')
plt.ylabel('Highest Yearly Earning')
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## india and united state this country are top highest monthly and highest yearly earning country in youtube

# 22) Which category shows the most significant growth in subscribers over the observed years? 

```python
agg_df = df.groupby(['created_year', 'category'])['subscribers'].sum().reset_index()

# Plot the area plot
plt.figure(figsize=(12, 8))
sns.lineplot(x='created_year', y='subscribers', hue='category', data=agg_df, marker='o', estimator=None)
plt.fill_between(agg_df['created_year'], agg_df['subscribers'], color='skyblue', alpha=0.3)
plt.title('Subscriber Growth Over Time by Category')
plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Channels centered around music and comedy consistently attract more subscribers over time, indicating the enduring popularity and appeal of these genres among YouTube audiences.
